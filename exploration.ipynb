# Exploration Phase

### Research Goal

We aim to analyze the possible factors determining consumer average spending amount, such as gender, geographical location, product type, payment method, discount or promotion, etc. We gathered shopping trend datasets to make further prediction.

Research Question and Hypotheses R1 Which factors influence the consumer average spending amount? For each factor independently: H0: this factor has a significant effect on consumer average spending amount.

Potential factors: a. Demographics: age, gender (categorical), income level (continuous) b. Geographic: country/regions (categorical) c. Individual shopping behavior: product type (categorical), payment method (categorical) d. External: Discount (binary: Yes=1, No=0), holiday season (binary: Yes=1, No=0)

### Initial Exploration

For one of the e-commerce dataset, we worked on a regression analysis to see how factors like customer age and product categories affect the total purchase amount. The process was interesting, but the results were not as strong as we hoped. Customer age and product categories were used as predictors and fitted a linear regression model. 

- The R-squared value was only 0.003, meaning the model explains just 0.3% of the variation in total purchase amount. 

- Customer Age: There is a significant positive relationship with total purchase amount (p < 0.001). For every extra year of age, the purchase amount increases by about 5.31 units.

- Product Categories: The Home category showed a positive effect (p = 0.045); Clothing and Electronics had no significant impact on spending.

While the analysis gave some insights, the overall model performance was very weak. Because of this, We decided not to continue with the regression. The current variables don’t seem enough to explain spending behavior.
